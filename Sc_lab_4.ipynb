```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Binary Step Function
def binary_step(x):
    return np.where(x >= 0, 1, 0)

# 2. Bipolar Sigmoidal Function
def bipolar_sigmoid(x):
    return (1 - np.exp(-x)) / (1 + np.exp(-x))

# 3. Binary Sigmoidal Function (Logistic Sigmoid)
def binary_sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 4. ReLU Function
def relu(x):
    return np.maximum(0, x)

def plot_activation_functions():
    x = np.linspace(-10, 10, 400)

    plt.figure(figsize=(12, 8))

    # Plot Binary Step
    plt.subplot(2, 2, 1)
    plt.plot(x, binary_step(x), label="Binary Step")
    plt.title("Binary Step Function")
    plt.grid(True)
    plt.ylim(-0.2, 1.2)
    plt.xlabel("x")
    plt.ylabel("f(x)")

    # Plot Bipolar Sigmoidal
    plt.subplot(2, 2, 2)
    plt.plot(x, bipolar_sigmoid(x), label="Bipolar Sigmoid", color='orange')
    plt.title("Bipolar Sigmoidal Function")
    plt.grid(True)
    plt.ylim(-1.2, 1.2)
    plt.xlabel("x")
    plt.ylabel("f(x)")

    # Plot Binary Sigmoidal
    plt.subplot(2, 2, 3)
    plt.plot(x, binary_sigmoid(x), label="Binary Sigmoid", color='green')
    plt.title("Binary Sigmoidal Function")
    plt.grid(True)
    plt.ylim(-0.2, 1.2)
    plt.xlabel("x")
    plt.ylabel("f(x)")

    # Plot ReLU
    plt.subplot(2, 2, 4)
    plt.plot(x, relu(x), label="ReLU", color='red')
    plt.title("ReLU Function")
    plt.grid(True)
    plt.ylim(-1, 10)
    plt.xlabel("x")
    plt.ylabel("f(x)")

    plt.tight_layout()
    plt.show()

# Run the plot function
plot_activation_functions()
```
